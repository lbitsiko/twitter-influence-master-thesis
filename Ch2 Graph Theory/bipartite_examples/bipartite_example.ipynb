```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
# %matplotlib widget
```

```python
fig, ax = plt.subplots()
ax.axis('off')
dict_of_lists_graph = {1:[4,5],
                       2:[6,7],
                       3:[5,6],
                       4:[1],
                       5:[1,3],
                       6:[2,3]}
color_map = 3* ['tab:blue'] + 4 *['tab:orange']
G = nx.from_dict_of_lists(dict_of_lists_graph)
spring_layout_for_G = nx.spring_layout(G, seed=70) # position of nodes not random
nx.draw_circular(G, node_color = color_map, with_labels = True)#,pos=spring_layout_for_G,with_labels=False)
fig.savefig('./bipartite_example.png')
```
