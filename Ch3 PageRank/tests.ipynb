```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
# %matplotlib widget
```

```python
fig, ax = plt.subplots()
# ax.axis('off')
G = nx.Graph()  # an undirected graph object

for i in range(6): G.add_node(i + 1)

G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(3, 1)
G.add_edge(3, 2)
G.add_edge(4, 5)
G.add_edge(4, 6)
G.add_edge(5, 4)
G.add_edge(5, 6)
G.add_edge(6, 4)
G.add_edge(3, 5)
spring_layout_for_G = nx.spring_layout(G, seed=100) # position of nodes not random
nx.draw_circular(G)#, pos=spring_layout_for_G)
# fig.savefig('./plots/Simple_pagerank_G.png', format='PNG')
```

```python
fig2, ax = plt.subplots()
# ax.axis('off')
G = nx.DiGraph()  # a directed graph object
for i in range(6): G.add_node(i + 1)

G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(3, 1)
G.add_edge(3, 2)
G.add_edge(4, 5)
G.add_edge(4, 6)
G.add_edge(5, 4)
G.add_edge(5, 6)
G.add_edge(6, 4)
G.add_edge(3, 5)
nx.draw_circular(G)#, pos=spring_layout_for_G)
# fig2.savefig('./plots/Simple_pagerank_DiGraph.png', format='PNG')
```

```python
pageranks = nx.pagerank(G)
```

```python
pagerank_vals = [pageranks[key] for key in pageranks.keys()]
```

```python
fig, ax = plt.subplots()
nx.draw_circular(G, node_color = pagerank_vals, cmap=plt.cm.viridis)#, pos=spring_layout_for_G)
```

```python
G = nx.generators.random_graphs.binomial_graph(1000, 0.3, directed=False)
# pos = nx.get_node_attributes(G, "pos")
```

```python
fig, ax = plt.subplots()


pos = nx.spring_layout(G, seed=100)

pageranks = nx.pagerank(G)
pagerank_vals = [pageranks[key] for key in pageranks.keys()]

# nx.draw_networkx_edges(G, pos, alpha=0.005)
nx.draw_networkx_nodes(
    G,
    pos,
    # nodelist=list(p.keys()),
    node_size=80,
    node_color=pagerank_vals,
    cmap=plt.cm.Reds_r,
)


# nx.draw_circular(G, node_color = pagerank_vals, cmap = plt.cm.Blues, node_size = 8, edge_)#, pos=spring_layout_for_G)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(1.5*6.4, 4.8), tight_layout=True)
# fig, ax = plt.subplots()
ax.plot(range(1000), pagerank_vals, '.-', alpha = 0.5, lw = 0.5)
ax.hlines(np.mean(pagerank_vals), 0, 1000, color = 'red')
```
